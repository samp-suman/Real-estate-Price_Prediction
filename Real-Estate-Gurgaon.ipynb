````python


# Define a function to categorize values
def categorize_age(age):
    if pd.isna(age):
        return 'undefined'
    elif 'Old' in age  in age or pd.to_datetime(age, errors='coerce') < pd.to_datetime('Aug 2023', format='%b %Y', errors='coerce'):
        return 'constructed'
    else:
        return 'underConstruction'

data['status'] = data['agePossession'].apply(categorize_age)
data[['status', 'agePossession']]
````

````python
def calculate_average_rating(row):
  ratings = []
  # print(row, type(row))
  if type(row)!=float:
    for value in row.strip('\[|\]').split(', '):
      # print(value)
      rating = int(value.split(' out')[0][-1])
      ratings.append(rating)
    if ratings:
      return sum(ratings) / len(ratings)
    else:
      return None


# Apply the function to create a new column with average ratings
data['avgRating'] = data['rating'].apply(calculate_average_rating)
````

* Created a rating columns based on averaging all different parameter ratings

#### FurnishDetails

````python
import ast

# Define the keys_list
keys_list = ['Modular Kitchen', 'Bed', 'Washing Machine', 'TV', 'Stove', 'Sofa', 'Fridge', 'Microwave', 'Dining Table', 'Water Purifier', 'Wardrobe', 'Curtains', 'Chimney', 'AC', 'Light', 'Geyser', 'Exhaust Fan', 'Fan']

# Function to create a dictionary from the input string
def create_furnishing_dict(input_string):
    item_dict = {key: 0 for key in keys_list}
    if type(input_string)!=float:
      input_list = ast.literal_eval(input_string)
      for item in input_list:
          v = item.split(' ', 1)[0]
          key = item.split(' ', 1)[1]
          item_dict[key]=v
      return item_dict
    else:
      return None
data['furnishDetailsDict'] = data.furnishDetails.apply(create_furnishing_dict)
````

````python
def get_furnish(x):
  if x and type(x)!=float:
    if col in x:
      return x[col]
    else:
      return 0
  else:
    return None

keys_list = ['Modular Kitchen', 'Bed', 'Washing Machine', 'TV', 'Stove', 'Sofa', 'Fridge', 'Microwave', 'Dining Table', 'Water Purifier', 'Wardrobe', 'Curtains', 'Chimney', 'AC', 'Light', 'Geyser', 'Exhaust Fan', 'Fan']
for col in keys_list:
  data[col] = data.furnishDetailsDict.apply(get_furnish)
````
